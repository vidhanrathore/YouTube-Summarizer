```python
import yt_dlp
import os

def get_youtube_captions(video_url, language='en'):
    try:
        # First, extract video info to check available subtitles
        with yt_dlp.YoutubeDL({'quiet': True}) as ydl:
            info_dict = ydl.extract_info(video_url, download=False)
            subtitles = info_dict.get('subtitles', {})
            auto_captions = info_dict.get('automatic_captions', {})

            video_id = info_dict.get("id", None)

        # Decide where to pull from: manual or auto
        caption_source = 'subtitles' if language in subtitles else 'automatic_captions' if language in auto_captions else None
        if not caption_source:
            return None  # No captions available

        ydl_opts = {
            'skip_download': True,
            'writesubtitles': caption_source == 'subtitles',
            'writeautomaticsub': caption_source == 'automatic_captions',
            'subtitleslangs': [language],
            'subtitlesformat': 'vtt',
            'outtmpl': '%(id)s.%(ext)s',
            'quiet': True
        }

        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            ydl.download([video_url])

        subtitle_file = f"{video_id}.{language}.vtt"
        if os.path.exists(subtitle_file):
            with open(subtitle_file, 'r', encoding='utf-8') as f:
                vtt_data = f.read()
            os.remove(subtitle_file)
            return vtt_data
        else:
            return None

    except Exception as e:
        print(f"Error: {e}")
        return None
```

```python
yt = "https://www.youtube.com/watch?v=Qa6csfkK7_I"
result = get_youtube_captions(yt)
print(result)
```

```python
from youtube_transcript_api import YouTubeTranscriptApi
from youtube_transcript_api.formatters import TextFormatter
import re

def get_youtube_id(url):
    # Extract video ID from URL
    match = re.search(r"(?:v=|\/)([0-9A-Za-z_-]{11})", url)
    return match.group(1) if match else None

def get_transcript_api_captions(video_url, lang='en'):
    video_id = get_youtube_id(video_url)
    print(video_id)
    try:
        transcript = YouTubeTranscriptApi.get_transcript(video_id, languages=[lang])
        print(transcript)
        text = ""
        for dic in transcript:
            text += dic["text"]
            text += " "
        return text
    except Exception as e:
        return f"Error: {e}"

result = get_transcript_api_captions(yt)
print(result)
```
